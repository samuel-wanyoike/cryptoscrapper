```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import pandas as pd
import json
from my_config import API_KEY

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'15',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': API_KEY,
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = response.json()

except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)

crypto_data = data['data']
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.json_normalize(crypto_data)
df['timestamp'] = pd.to_datetime('now')
```

```python
def api_trigger():
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
    'start':'1',
    'limit':'15',
    'convert':'USD'
    }
    headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': API_KEY,
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = response.json()

    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)

    crypto_data = data['data']
    df2 = pd.json_normalize(crypto_data)
    df2['timestamp'] = pd.to_datetime('now')
    # append df2 to existing global df (create df if it doesn't exist)
    global df
    if 'df' in globals():
        df = pd.concat([df, df2], ignore_index=True)
    else:
        df = df2.copy()
```

```python
import os
from time import time
from time import sleep
from my_config import API_KEY  # ensure API_KEY is available in this cell's namespace

for i in range(333):
    api_trigger()
    print('api trigger completed')
    sleep(60) #sleep for 1 minute
exit()
```
